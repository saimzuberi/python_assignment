```python
#import pandas 
import pandas as pd
#import csv files from path
from pathlib import Path
#define file paths
csvpath = Path("sales_data.csv", index_col = 'Line_Item_ID')
csvpath_1 = Path("menu_data.csv")

#import file in dataframe sales_df and menu_df
sales_df = pd.read_csv(csvpath)
menu_df = pd.read_csv(csvpath_1)

#set index for Menu_item and item in respected files. 
sales_df.set_index('Menu_Item', inplace = True)
menu_df.set_index('item', inplace = True)

#merge both files based on index values to concatnate cols.
merged_data_cols = pd.merge(
    sales_df, menu_df, left_index=True, right_index = True)
##print(merged_data_cols)

#Data Cleaning with deletion w.r.t to cols not needed
del merged_data_cols ["Line_Item_ID"]
del merged_data_cols ["Date"]
del merged_data_cols ["Credit_Card_Number"]
del merged_data_cols ["description"]
del merged_data_cols ["category"]
#print(merged_data_cols.head())
#merged_data_cols.sum()

# Reset index to access index col for group by statments
new_df = merged_data_cols.reset_index()
#print(new_df.head())

# perform Calculations for Revenue
new_df['revenue'] = new_df['Quantity']*new_df['price']
#new_df.head()
# perform calculations for Cost of Goods Sold
new_df['cog']= new_df['Quantity']*new_df['cost']
#new_df.head()
# perform calculations for profit
new_df['profit'] = new_df['revenue']-new_df['cog']

# Bring data into new dataframe
sale_count = new_df.groupby('index').sum()

# Clean Data
del sale_count ["price"]
del sale_count ['cost']
# Display Output
sale_count.head(11)

# Set output file name
output_path = 'Financial_Analysis.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    # Write Financial Analysis to the output file, converting parameters to string
    file.write(f"FINANCIAL ANALYSIS \n")
    file.write("-------------------------------------------------------------------------------------\n")
    file.write(f"{sale_count}")

```
