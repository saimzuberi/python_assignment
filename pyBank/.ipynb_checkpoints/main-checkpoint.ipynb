```python
# importing csv module
import csv

# csv file name
filename = "budget_data.csv"

""" ------------------------------------------------------- """

# initializing the titles and rows list
# initializing parameters for CSV File
fields = []
rows = []

# initializing parameters for month list
month = []
months = ""
total_months = 0

# initializing parameters for profit/loss list 
profit_loss = []
total_pnl = 0

# initializing parameters for list for average change 
average_change_pnl = []
average_pnl = 0
total_average_pnl = 0
total_counts = 0
average_diff = 0
total_change_pnl = 0

""" ------------------------------------------------------- """

# reading csv file
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
    
    # extracting field names through first row
    fields = next(csvreader)

    # extracting each data row one by one
    for row in csvreader:
        rows.append(row)

    # get total number of rows
    print("Total no. of rows: %d"%(csvreader.line_num))

""" ------------------------------------------------------- """
    
# printing the field names
print('Field names are:' + ', '.join(field for field in fields))

# printing first 5 row
'''print('\nFirst 5 rows are:\n')
for row in rows[:5]:
    # parsing each column of a row
    for col in row:
        print("%10s"%col,end=" "),
    print('\n')'''

# get total number of months by reviewing the data file and identifying header and months
print("Total no. of months: %d"%(csvreader.line_num-1))

""" ------------------------------------------------------- """

# adding profit and loss to list profit_loss and adding months to month list
x = 0
for row in rows:
    pnl = int(row[1])
    profit_loss.append(pnl)
    months = row[0]
    if x > 0:
        month.append(months)
    x += 1
#print(month)
#print("\n-------\n")

# finding total profit 
    
for pnl in profit_loss:
    total_pnl += pnl
    total_months += 1
    
    
print(f"the total net profit/loss for the bank is $ {total_pnl}")

""" ------------------------------------------------------- """

# The average of the changes in Profit/Losses over the entire period

length = len(profit_loss)
i = 1
while i < length:
    average_pnl = (int(profit_loss[i])- int(profit_loss[i-1]))
    average_change_pnl.append(average_pnl)
    i += 1
    
for average_pnl in average_change_pnl:
    total_change_pnl += average_pnl
    total_counts += 1

# Average difference in profit/loss  

average_diff = round(total_change_pnl/total_counts,2)
print(f"The average of the changes in Profit/Losses over the entire period is ${average_diff}")

""" ---------------------------------------------------------- """"

# creating a dictonary with months and average difference in profit and loss 

keys = month
values = average_change_pnl
new_dictionary = dict(zip(keys, values))

#finding maximum profit/loss and coresponding month

temp = max(new_dictionary.values())
res = [key for key in new_dictionary if new_dictionary[key] == temp]
print("Greatest Increase in Profits " + str(res) + str(temp))

#finding minimum profit/loss and coresponding month. 

temp_1 = min(new_dictionary.values())
res_1 = [key for key in new_dictionary if new_dictionary[key] == temp_1]
print("Greatest decrease in losses: " + str(res_1) + str(temp_1))

""" ----------------------------------------------------- """

# Set output file name
output_path = 'Financial_Analysis.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    # Write Financial Analysis to the output file, converting parameters to string
    file.write(f"FINANCIAL ANALYSIS \n")
    file.write("-------------------------------------------------------------------------------------\n")
    file.write(f"Total no. of rows: %d"%(csvreader.line_num) + "\n")
    file.write(f"Field names are:" + ', '.join(field for field in fields) + "\n")
    file.write(f"Total no. of months: %d"%(csvreader.line_num-1) + "\n")
    file.write(f"The average of the changes in Profit/Losses over the entire period is ${average_diff} \n")
    file.write(f"Greatest Increase in Profits " + str(res) + str(temp) + "\n")
    file.write(f"Greatest decrease in losses: " + str(res_1) + str(temp_1) + "\n")
```
